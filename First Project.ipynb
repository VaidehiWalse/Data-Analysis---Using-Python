```python
plt.figure(figsize = (5,5))
ax = sns.countplot(x = 'Gender',data = df )#x= x-axis and y will the count of the that gender 
plt.title('Gender Data')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
>> Calculating the number of sales done by Male & Female
```

```python
df.groupby(['Gender'],as_index= False)['Amount'].sum().sort_values(by='Amount', ascending = False)
```

```python
>> Plot the graph for the above
```

```python
sales_gender = df.groupby(['Gender'], as_index= False)['Amount'].sum().sort_values(by='Amount',ascending = False)
ax= sns.barplot(x='Gender',y = 'Amount',data = sales_gender)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
From the above graphs we can see that most of the buyers are females and even the purchases are maximum done by females
```
